### Choosen Decay action: Stretched Exponential Decay

The decay function is chosen is such a manner there is more dwell time in initial part of episodes (epsilon > 1) to ensure agent explores quite a lot before getting greedy<br>Until 7000 episodes (1/4th) the epsilon value is greater than 1<br>Similarly after 50% of episode times, the epsilon is <0.1<br>Shape of decay function is shown in graph below<br> For more details on the decay function, visit this page https://medium.com/analytics-vidhya/stretched-exponential-decay-function-for-epsilon-greedy-algorithm-98da6224c22f

```python
#Epsilon function
def epsilon(time):
    standardized_time=(time-0.35*Episodes)/(0.1*Episodes)
    cosh=np.cosh(math.exp(-standardized_time))
    epsilon=1.1-(1/cosh+(time*0.1/Episodes))
    return epsilon
```

      - Initialise your input and output batch for training the model
      - Calculate the target Q value for each sample: reward + gamma*max(Q(s'a,))
      - Get Q(s', a) values from the last trained model
      - Update the input batch as your encoded state and output batch as your Q-values
      - Then fit your DQN model using the updated input and output batch.

```python
class DQNAgent:
    def __init__(self, state_size, action_size):
        # Define size of state and action
        self.state_size = state_size
        self.action_size = action_size

        # Write here: Specify you hyper parameters for the DQN
        # we tried several hyper parameter combinations fr learning rate and discount factor
        # the optimal values that gave maximum total reward per episode were chosen as below
        self.discount_factor = 0.2
        self.learning_rate = 0.001       
        self.epsilon_max = 1
        self.epsilon_decay = 0.01
        self.epsilon_min = 0
        
        self.batch_size = 32        
        # create replay memory using deque
        self.memory = deque(maxlen=2000)

        # create main model and target model
        self.model = self.build_model()

    # approximate Q function using Neural Network
    def build_model(self):
        model = Sequential()
        # Write your code here: Add layers to your neural nets       
        # hidden layers
        model.add(Dense(32, input_dim=self.state_size, activation='relu', kernel_initializer='he_uniform'))
        model.add(Dense(32, activation='relu', kernel_initializer='he_uniform'))

        # the output layer: output is of size num_actions
        model.add(Dense(self.action_size+1, activation='relu', kernel_initializer='he_uniform'))
        model.compile(loss='mse',optimizer=Adam(lr=self.learning_rate))
        model.summary
        return model



    def get_action(self, state,possible_actions,time):
    # Write your code here:
    # get action from model using epsilon-greedy policy
    # this function seeks current episode number and possible actions as input
    # this function output the index of chosen action
            if np.random.rand() <=epsilon(time):
            # explore: choose a random action from all possible actions
            
                random.seed(datetime.now()) #seeded with current time to make a choice nearest to true random
                return random.choice(possible_actions)
            else:
            # choose the action with the highest q(s, a)
            # the first index corresponds to the batch size, so
            # reshape state to (1, state_size) so that the first index corresponds to the batch size
                state = env.state_encod_arch1(state)
                state=state.reshape(1, self.state_size)
                q_value = self.model.predict(state)
                return possible_actions[np.argmax(q_value[0][possible_actions])]
        
        
        
        


    def append_sample(self, state, action, reward, next_state):
    # Write your code here:
    # save sample <s,a,r,s'> to the replay memory
        self.memory.append((state, action, reward, next_state))
    
    
    # pick samples randomly from replay memory (with batch_size) and train the network
    def train_model(self):
        
        if len(self.memory) > self.batch_size:
            # Sample batch from the memory
            mini_batch = random.sample(self.memory, self.batch_size)
            update_output = np.zeros((self.batch_size, self.state_size))
            update_input = np.zeros((self.batch_size, self.state_size))
            
            actions,rewards=[],[]
           
            
            for i in range(self.batch_size):
                state, action, reward, next_state = mini_batch[i]
                update_input[i] = env.state_encod_arch1(state)
                actions.append(action)
                rewards.append(reward)
                update_output[i] = env.state_encod_arch1(next_state)
                
                # Write your code from here
                # 1. Predict the target from earlier model
           
            target = self.model.predict(update_input)     
                
                # 2. Get the target for the Q-network
            target_qval = self.model.predict(update_output)    
                
                #3. Update your 'update_output' and 'update_input' batch

            for i in range(self.batch_size):
                prior_record=target[i][actions[i]]
                target[i][actions[i]] += self.learning_rate*(rewards[i] + self.discount_factor * np.max(target_qval[i]) -prior_record)    
                
        # 4. Fit your model and track the loss values
            self.model.fit(update_input, target, batch_size=self.batch_size, epochs=1, verbose=0)

    def save_model_weights(self, name):
        self.model.save_weights(name)
        
```

### DQN block

```python
start_time=time.time()

# Call the environment
env=CabDriver()
print("Environment Initialized")
# Call all the initialised variables of the environment

state=env.state_init
state_size=len(env.state_encod_arch1(state))
action_size=len(env.action_space)

#Call the DQN agent
agent=DQNAgent(state_size,action_size)
print("Agent Initialized")
if data_exist:
    #in case the training is resumed from a previous run, saved model weights are reloaded
    agent.model.load_weights('./model_weights.h5')
    print('Model loaded from existing data')
    episode=len(rewards_tracked)
    print("Training from episode:",episode)

for episode in range(Episodes):
    
    #reset environment at start of each episode and assign new state 
    state=env.reset()[2]
      
    total_trip_time=0  #initalize trip time to zero at beginning of every episode
    total_reward=0 #initialize total reward to zero at beginning of every episode
    while total_trip_time<=30*24:
        
        # 1. Pick epsilon-greedy action from possible actions for the current state
        possible_actions=env.requests(state)
        chosen_action=agent.get_action(state,possible_actions[0],episode)
        
        # 2. Evaluate your reward and next state
        next_state=env.next_state_func(state, chosen_action, Time_matrix)
        reward,trip_time=env.reward_func(state,chosen_action, Time_matrix)
        total_reward+=reward #cumulate reward in total rewards per episode
        
        #track states
        if state in States_track and chosen_action in States_track[state]:
            one_hotcoded_state = env.state_encod_arch1(state) #convert state to one hot encoded vector
            one_hotcoded_state=one_hotcoded_state.reshape(1, state_size) #reshape vector to suit the NN architecture
            q_value=agent.model.predict(one_hotcoded_state) #agent predicts the Q values for given state
            q_value=q_value[0][chosen_action] #we chose the action with maximum Q value
            #store chosen action and q value corresponding to the state for future tracking
            States_track[state][chosen_action].append(q_value)
               
        
        # 3. Append the experience to the memory
        agent.append_sample(state,chosen_action, reward, next_state)
        total_trip_time+=trip_time #cumulate trip time
        state = next_state #make next state as current state for next trip
        
        # 4. Train the model by calling function agent.train_model
        agent.train_model()
    
    rewards_tracked.append(total_reward) #append total reward at end of each episode
   
    stdout.write('Episode: {:,} \r'.format(episode+1))
    stdout.flush()
    
    
    if ((episode+1) % threshold) == 0:   #every 500th episode
        
        save_obj(States_track,'States_tracked')
        save_obj(rewards_tracked,'Rewards_tracked')
            
   
    # save model weights
    # every 500th episode
    if (episode+1) % policy_threshold == 0:  
        agent.save_model_weights(name="model_weights.h5")
    
   
  
    
print('Episode 'f"{episode+1:,d} completed ")
elapsed_time = time.time() - start_time
print('Time lapsed:',elapsed_time)

save_obj(States_track,'States_tracked')   
agent.save_model_weights(name="model_weights.h5")
save_obj(rewards_tracked,'Rewards_tracked')
print('Saved Model, Rewards and States tracked')
    
```
