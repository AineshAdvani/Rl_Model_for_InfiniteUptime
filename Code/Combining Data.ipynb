```python
import pandas as pd
import glob
```

```python

# Get a list of all Excel files in the folder
excel_files = glob.glob('C:\\Users\\advan\\OneDrive\\Desktop\\Infinite Uptime\\DATA\\Alerts\\May\\*.xlsx')  # Replace 'path_to_folder' with the actual folder path

# List to store DataFrames for each file
dataframes = []

# Read each Excel file into a DataFrame and store in the list
for file in excel_files:
    df = pd.read_excel(file)
    dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(dataframes, ignore_index=True)

# Save the merged DataFrame to a new Excel file
merged_df.to_excel('C:\\Users\\advan\\OneDrive\\Desktop\\Infinite Uptime\\DATA\\May.xlsx', index=False)
```

Cleaning the Data

```python
may = pd.read_excel('C:\\Users\\advan\\OneDrive\\Desktop\\Infinite Uptime\\DATA\\May.xlsx')
new_may=may.drop(['machine_name','monitor_name','timestamp','CoE Remarks if any','DRS ID (Existing / Newly created)'],axis=1)
new_may
```

```python
# getting config data

config = pd.read_excel('C:\\Users\\advan\\OneDrive\\Desktop\\Infinite Uptime\\DATA\\machine_config_3July2023.xlsx')
config
```

```python
new_config = config[['monitor_id','plant_name_y','mounting_condition','speed','speed_type','m_category']]
new_config
```

```python
new_config.rename(columns={'plant_name_y': 'plant_name'}, inplace=True)
new_config
```

```python
train = new_may.merge(new_config, on=['monitor_id', 'plant_name'], how='left')
train
```

```python
train['plant_id'].min()
```

```python
v1_train = train.dropna(subset=['CoE Validation as per SOP'])
v1_train
```

```python
v2_train = v1_train[v1_train['CoE Validation as per SOP'] != 'UNKNOWN']
v2_train
```

```python
# modifying customer feedback for training purpose
import numpy as np
v2_train['Customer Validation'] = v2_train['Customer Validation'].apply(lambda x: 0 if x == 'FP' else 1)
v2_train['CoE Validation as per SOP'] = v2_train['CoE Validation as per SOP'].apply(lambda x: 1 if x == True else 0)
v2_train
```

```python
continuous_columns = ['magnitude','Increase in magnitude (multiplier)','speed']

for col in v2_train.columns:
    if col not in continuous_columns:
        v2_train[col] = pd.factorize(v2_train[col])[0]
```
